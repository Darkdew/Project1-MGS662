```python
import pandas as pd
import numpy as np
```

```python
df_text= pd.read_excel(r'D:\Syllabus and guidelines\662- Optimization Methods in ML\Project1\train_animal\animal_text_featuresand_labels.xlsx')
```

```python
df_image= pd.read_excel(r'D:\Syllabus and guidelines\662- Optimization Methods in ML\Project1\train_animal\animal_imagefeatures.xlsx')
```

```python
list_filename=[]
list_final=[]
for rows1 in range(len(df_text)):
    for rows2 in range(len(df_image)):
        if df_text.iloc[rows1][0]==df_image.iloc[rows2][0]:
            list1= list(df_text.iloc[rows1][1:].values)
            list2= list(df_image.iloc[rows2][1:].values)
            list_combined= list1+ list2
            list_final.append(list_combined)
            list_filename.append(df_text.iloc[rows1][0])
```

```python
df_textandimage_features= pd.DataFrame(list_final)
df_textandimage_features.insert(loc=0, column= "FileName", value= list_filename)
df_textandimage_features.set_axis([*df_textandimage_features.columns[:-1], 'Label'], axis=1, inplace=False)
```

```python
df_textandimage_features.to_excel(r'D:\Syllabus and guidelines\662- Optimization Methods in ML\Project1\train_animal\animal_combined_featuresand_labels.xlsx', index= False)
```
