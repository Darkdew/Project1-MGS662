```python
# example of loading an image with the Keras API
from keras.preprocessing.image import load_img
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import array_to_img
import pandas as pd
```

# TREE

```python
# load the excel sheet with file names of each of the image file
#class_1
df_tree_class1= pd.read_excel('/content/drive/MyDrive/Project1/train_tree/tree_Class_1.xlsx')

#creating a list to store all 75 features for all the files in this class
listFinal=[]

#looping on all the rows of the file names in the class1 category 
for i in range(len(df_tree_class1)):
  list1=[]
  
  #parameterizing the location of each file with its unique name
  a=str('/content/drive/MyDrive/Project1/train_tree/Class_1/')
  b= str(df_tree_class1.iloc[i, 0])
  file_loc_str=  a+b

  #loading each of the image in the loop one at a time
  img = load_img(path=file_loc_str, target_size=(5,5))
  
  # converting this img to array
  img_array = img_to_array(img)

  #reshaping the 3 channel img_array to 1 dimension with 5*5*3
  reshaped_array= img_array.reshape(1,75)
  
  #converting the array to a list and appending to a final list
  list1= list(reshaped_array[0])
  
  listFinal.append(list1)
```

```python
#defining column names for each of the 75 features extracted
column_list=[]
for k in range(1,76):
  col_name= "feature"+ str(k)
  column_list.append(col_name)

print(len(column_list))
```

```python
#defining a new features dataframe with each of the 75 features
#of a particular image as rows
df_features= pd.DataFrame(listFinal, columns=column_list )
```

```python
#lets check out what we got
df_features.head()
```

```python
#creating a new list of all the filenames in class1
File_name_list= list(df_tree_class1.iloc[:, 0])
```

```python
#inserting the File_name_list as column0 in the df_features dataframe

df_features.insert(loc=0, column= "FileName", value= File_name_list )
```

```python
df_features.shape
```

```python
#inserting the label column with class 1
label_list= []
for values in range(len(df_features)):
  label_list.append(1)

df_features.insert(loc=76, column= "Class_label", value= label_list )
```

```python
# load the excel sheet with file names of each of the image file
#class_0
df_tree_class0= pd.read_excel('/content/drive/MyDrive/Project1/train_tree/tree_Class_0.xlsx')

#creating a list to store all 75 features for all the files in this class
listFinal=[]

#looping on all the rows of the file names in the class1 category 
for i in range(len(df_tree_class0)):
  list1=[]
  
  #parameterizing the location of each file with its unique name
  a=str('/content/drive/MyDrive/Project1/train_tree/Class_0/')
  b= str(df_tree_class0.iloc[i, 0])
  file_loc_str=  a+b

  #loading each of the image in the loop one at a time
  img = load_img(path=file_loc_str, target_size=(5,5))
  
  # converting this img to array
  img_array = img_to_array(img)

  #reshaping the 3 channel img_array to 1 dimension with 5*5*3
  reshaped_array= img_array.reshape(1,75)
  
  #converting the array to a list and appending to a final list
  list1= list(reshaped_array[0])
  
  listFinal.append(list1)


#defining column names for each of the 75 features extracted
column_list=[]
for k in range(1,76):
  col_name= "feature"+ str(k)
  column_list.append(col_name)


#defining a new features dataframe with each of the 75 features
#of a particular image as rows
df_features_0= pd.DataFrame(listFinal, columns=column_list )

#creating a new list of all the filenames in class1
File_name_list= list(df_tree_class0.iloc[:, 0])

#inserting the File_name_list as column0 in the df_features dataframe

df_features_0.insert(loc=0, column= "FileName", value= File_name_list )


#lets see how our extracted features along with respective file names looks
df_features_0.head()


#inserting the label column with class 0
label_list= []
for values in range(len(df_features_0)):
  label_list.append(0)

df_features_0.insert(loc=76, column= "Class_label", value= label_list )
```

```python
#lets see how our extracted features along with respective file names looks
df_features_0.head()
```

```python
#appending the 2 dataframes vertically
#this will give us the complete dataset of trees data with file name, 75 features and respective label
df_img_features=df_features.append(df_features_0, ignore_index=True)
```
